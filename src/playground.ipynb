```python
scope = """
        user-read-private 
        user-read-playback-state 
        playlist-modify-public 
        user-library-modify 
        user-library-read 
        playlist-read-private 
        playlist-modify-private
        """

sp = spotipy.Spotify(auth_manager=SpotifyOAuth(scope=scope))
```

```python
import pandas as pd
LIBRARY_FILEPATH = "../data/library.csv"
ARTIST_FILEPATH = "../data/artists.csv"
```

```python

library = pd.read_csv(LIBRARY_FILEPATH)
song_keys = library.keys()
library = library[song_keys[1:]]
library.head()
```

```python
artists = pd.read_csv(ARTIST_FILEPATH)
artists.head()
```

```python

library["genres"] = ""
library.head()
```

```python

import ast


for idx in range(library.shape[0]):
    song_artists = ast.literal_eval(library["artists"][idx])
    song_genres = set()
    for song_artist in song_artists:
        genre_list_unconverted = ((artists.loc[artists["uri"] == song_artist])["genres"]).to_list()
        if len(genre_list_unconverted) == 0:
            continue
        genre_list = ast.literal_eval(genre_list_unconverted[0])
        song_genres.update(genre_list)
    
    library["genres"][idx] = list(song_genres)

library.head()
```
